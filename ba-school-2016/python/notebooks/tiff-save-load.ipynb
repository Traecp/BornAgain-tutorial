```python
import numpy as np
from matplotlib import pyplot
from PIL import Image
```

```python
x = np.linspace(-1, 1)
y = np.linspace(-1, 1)
z = [ [u + 1.0j*v for u in x ] for v in y]
f = np.abs(z)
```

```python
image = pyplot.imshow(f)
image.figure.savefig('figure.tiff', format='tif')
pyplot.show()
```

```python
fig = pyplot.figure(frameon=False)
#fig.set_size_inches(w,h)
ax = pyplot.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
ax.imshow(f)
fig.savefig('raw_image.tiff', format='tif')
```

```python
tiff_image = Image.open('raw_image.tiff')
pyplot.imshow(tiff_image)
pyplot.show()
```
